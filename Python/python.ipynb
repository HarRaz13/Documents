```python
import matplotlib.pyplot as plt

Name=['A','B','C','D','E']
Python=[20,30,17,37,32]
ADBMS=[23,40,21,37,12]

plt.bar(Name, ADBMS, color="red")
plt.xlabel("Names")
plt.ylabel("ADBMS")
plt.show()

plt.scatter(x=Name, y=Python, color="pink")
plt.xlabel("Names")
plt.ylabel("Python")
plt.show()
```
